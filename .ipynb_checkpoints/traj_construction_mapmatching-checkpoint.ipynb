```python
# Read the GPS CSV files
########################

import pandas as pd
import geopandas as gpd
from shapely import wkt

# Filename constants
dated = '2023-08-30'
dat = '30Aug'

# Read csv
kelowna_gps_01Aug = pd.read_csv(f'daywise_gps_kelowna_sep/kelowna_gps_{dated}.csv')

# Display
kelowna_gps_01Aug
```

```python
# Convert to geodataframe
#########################

import geopandas as gpd

# Create a geometry column from latitude and longitude and create gdf for Halifax
geometry = [Point(xy) for xy in zip(kelowna_gps_01Aug['longitude'], kelowna_gps_01Aug['latitude'])]
kelowna_gps_01Aug = gpd.GeoDataFrame(kelowna_gps_01Aug, geometry=geometry)
    
# Set the coordinate reference system (CRS) if needed
kelowna_gps_01Aug.set_crs(epsg=4326, inplace=True)  # WGS 84
display(kelowna_gps_01Aug)
```

```python
# Display the list of unique device_ids
#######################################

from cybstr.genera import prin

# Column name and count limit
col_name = 'registrationID'
limt = 20

# Group by registrationID and filter for those with more than 20 entries
filtered_device_ids = prin.print_unique_gps_ids_v1(kelowna_gps_01Aug, col_name, limt)

# Display the list of unique device_ids
print(filtered_device_ids[:50])
```

```python
# Data cleaning
###############

from cybstr.genera import preproc

# Column name and count limit
col_name = 'registrationID'
freq = 2

# Drop less frequent GPS columns
kelowna_gps_01Aug, num_thrice_occurrence_ids = preproc.drop_less_freq_ids(kelowna_gps_01Aug, col_name, freq)

# Calculate the number of such IDs
print(f"Number of registrationIDs occurring less than {freq} times in halifax_gps_{dat}: {num_thrice_occurrence_ids}")
```

```python
# Data cleaning - dropping rows with duplicate registrationID + Time combo
##########################################################################

from cybstr.genera import preproc

# Column name and count limit
gps_col_name = 'registrationID'
date_col_name = 'DateTime'

# Drop duplicates
kelowna_gps_01Aug, rows_dropped = preproc.drop_duplicate_gps(kelowna_gps_01Aug, gps_col_name, date_col_name)
kelowna_gps_01Aug = kelowna_gps_01Aug.to_crs(epsg=4326)

print(f"Number of rows dropped: {rows_dropped}")
display(kelowna_gps_01Aug)
```

```python
# Data cleaning - dropping GPS IDs with no movement (upto 3 decimal places in latitude/longitude)
#################################################################################################

from cybstr.genera import preproc

kelowna_gps_01Aug = preproc.remove_static_gps_entries(kelowna_gps_01Aug)
```

```python
# Basic analysis - Data duration
################################

# Calculate the start and end times
start_time_k = kelowna_gps_01Aug['DateTime'].min()
end_time_k = kelowna_gps_01Aug['DateTime'].max()

print(f"halifax_gps_{dat} data was collected from {start_time_k} to {end_time_k}.")
```

```python
# Loading East Kelowna shapefile
################################

import contextily as ctx

def display_shp_in_contextily_v1(path, crs, dim):
    # Read shapefile
    shp = gpd.read_file(path)
    
    # Set the Coordinate Reference System (CRS) if not already set (WGS84 - EPSG:4326 for lat/long)
    shp.to_crs(epsg=crs, inplace=True)
    
    # Plot the geometries on a basemap
    fig, ax = plt.subplots(figsize=(dim, dim))
    
    # Plot the roads (your data)
    shp.plot(ax=ax, color='blue', linewidth=1, alpha=0.3)
    
    # Use OpenStreetMap for the basemap
    ctx.add_basemap(ax, crs=shp.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)
    
    # Show the plot
    plt.show()

    return shp

# Constant: SHP path and CRS value
path_to_shp = 'C:/remote/20Jul/bc_gis/EKelowna.shp'
epsg = 4326

e_kelowna_shp = display_shp_in_contextily_v1(path_to_shp, epsg, 6)
```

```python
# Filtering for East Kelowna
############################

# Ensure the CRS of both GeoDataFrames matches
kelowna_gps_01Aug = kelowna_gps_01Aug.to_crs(e_kelowna_shp.crs)

# Perform spatial join to get points near the NS101 highway buffer
ekelowna_gps_01Aug = gpd.sjoin(
    kelowna_gps_01Aug, 
    e_kelowna_shp, 
    how="inner", 
    predicate="intersects"
)

# No need to redefine geometry column if it's already a GeoDataFrame
ekelowna_gps_01Aug = ekelowna_gps_01Aug.to_crs(epsg=3857)

# Plot the filtered points
fig, ax = plt.subplots(figsize=(10, 10))
ekelowna_gps_01Aug.plot(
    ax=ax, color='blue', markersize=5, label='kelowna_gps_01Aug_east'
)

# Add a basemap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

# Customize the plot
ax.set_title("kelowna_gps_01Aug_east")
plt.show()
```

<H3>traj_2025_06_03: Trajectory for all IDs in a day</H3>
Do HMM based map matching and construct trajectory from map matched points for all IDs in a day for Halifax

```python
# Sort based on DateTime
########################

ekelowna_gps_01Aug['DateTime'] = pd.to_datetime(ekelowna_gps_01Aug['DateTime'])
ekelowna_gps_01Aug = ekelowna_gps_01Aug.sort_values(by=['registrationID', 'DateTime'])
```

```python
# Assign trip_ids based on 1hr gap between consecutive GPS points
#################################################################

from cybstr.traject import trip_analysis

# Constants: column names and time limit trip
gps_col_name = "registrationID"
date_col_name = "DateTime"
timelimt = 3600

# Return and print the df
ekelowna_gps_01Aug = trip_analysis.calculate_trip_id(ekelowna_gps_01Aug, gps_col_name, date_col_name, timelimt)
ekelowna_gps_01Aug = ekelowna_gps_01Aug.to_crs(epsg=4326)
display(ekelowna_gps_01Aug.head())
```

```python
# Prepare the geodataframe for map matching
###########################################

# Ensure geometry column contains Point objects
import geopandas as gpd

ekelowna_gps_01Aug = gpd.GeoDataFrame(ekelowna_gps_01Aug, geometry=gpd.points_from_xy(ekelowna_gps_01Aug.longitude, ekelowna_gps_01Aug.latitude), crs="EPSG:4326")
ekelowna_gps_01Aug = ekelowna_gps_01Aug.reset_index()

# Check if the geometry column is valid
if isinstance(kelowna_gps_01Aug, gpd.GeoDataFrame):
    # Extract x (longitude) and y (latitude) from the geometry column
    ekelowna_gps_01Aug["x"] = ekelowna_gps_01Aug.geometry.x
    ekelowna_gps_01Aug["y"] = ekelowna_gps_01Aug.geometry.y
else:
    raise ValueError(f"halifax_gps_{dat} must be a GeoDataFrame with a valid geometry column.")

### wkelowna_gps_01Aug = wkelowna_gps_01Aug.drop(['level_0', 'index'], axis=1)
display(kelowna_gps_01Aug)
```

```python
# Initialize maps
#################

# Retrieve all lat/long and make a list
latitude_ekelowna_gps_01Aug = ekelowna_gps_01Aug["latitude"].to_list()
longitude_ekelowna_gps_01Aug = ekelowna_gps_01Aug["longitude"].to_list()

points_ekelowna_gps_01Aug = [(lat, lon) for lat, lon in zip(latitude_ekelowna_gps_01Aug[:30], longitude_ekelowna_gps_01Aug[:30])]

# Center map around mean latitude and longitude
loc = (np.mean(latitude_ekelowna_gps_01Aug[:30]), np.mean(longitude_ekelowna_gps_01Aug[:30]))
maps = visualization.Map(location=loc, zoom_start=15)
### maps.add_graph(points_halifax_gps_01Aug, plot_nodes=True)
### maps
```

```python
# Run the above methods to generate gdf and map
###############################################

from cybstr.traject import map_match

failed_trips = []

# Create a geometry column from latitude and longitude and create gdf for Halifax
geometry = [Point(xy) for xy in zip(ekelowna_gps_01Aug['longitude'], ekelowna_gps_01Aug['latitude'])]
ekelowna_gps_01Aug = gpd.GeoDataFrame(ekelowna_gps_01Aug, geometry=geometry)
    
# Set the coordinate reference system (CRS) if needed
ekelowna_gps_01Aug.set_crs(epsg=4326, inplace=True)  # WGS 84

# Call the function to map-match and create a GeoDataFrame
ekelowna_gps_01Aug, matched_trips_gdf, failed_trips = map_match.map_match_and_return_geodataframe_v4(
    df=ekelowna_gps_01Aug,
    log_name=f"ekelowna_gps_{dat}",
    graph=roads_kelowna_json, 
    day_col="DateTime", 
    gps_id_col="registrationID", 
    trip_id_col="trip_id", 
    interp_dist=50, 
    radius=90,
    max_trips = None
)
```
